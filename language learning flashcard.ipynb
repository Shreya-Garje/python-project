```python
import random
import time
import matplotlib.pyplot as plt
from gtts import gTTS
from playsound import playsound
import os
import uuid
import difflib

# Function to pronounce text using gTTS and playsound
def pronounce(text):
    try:
        filename = f"temp_{uuid.uuid4()}.mp3"
        tts = gTTS(text=text, lang='en')
        tts.save(filename)
        playsound(filename)
        os.remove(filename)
    except Exception as e:
        print(f"Pronunciation error: {e}")

# Flashcards categorized by topic
flashcards = {
    "vocabulary": [
        ('abundant', 'existing in large quantities', 'hard'),
        ('dull', 'boring', 'easy'),
        ('educated', 'learned', 'medium'),
        ('efficient', 'productive', 'medium'),
        ('fail', 'flunk', 'easy'),
        ('fast', 'quick', 'easy'),
        ('fresh', 'recent', 'easy'),
        ('gentle', 'mild', 'easy'),
        ('grow', 'expand', 'medium'),
        ('happy', 'joyful', 'easy'),
    ],
    "phrases": [
        ('break a leg', 'good luck', 'easy'),
        ('hit the nail on the head', 'get something exactly right', 'medium'),
        ('bite the bullet', 'endure a painful situation bravely', 'hard'),
        ('burn the midnight oil', 'work late into the night', 'medium'),
        ('kill two birds with one stone', 'solve two problems at once', 'easy'),
        ('under the weather', 'feeling sick', 'easy'),
        ('the ball is in your court', 'it is your turn to take action', 'easy'),
        ('through thick and thin', 'in good times and bad times', 'medium'),
        ('spill the beans', 'reveal a secret', 'easy'),
        ('once in a blue moon', 'something that happens rarely', 'easy'),
        ('put all your eggs in one basket', 'relying too much on one option', 'medium'),
        ("don't cry over spilled milk", "don't stress over something that can't be changed", 'hard'),
    ],
    "business": [
        ('ROI', 'Return on Investment', 'medium'),
        ('stakeholder', 'person with interest in a project', 'medium'),
    ]
}

# Progress tracker
progress = {
    "total_attempts": 0,
    "correct_answers": 0
}

# Review flashcards
def review_flashcards():
    for category, cards in flashcards.items():
        print(f"\n--- {category.capitalize()} ---")
        for word, translation, difficulty in cards:
            print(f"{word:<35}: {translation:<40} ({difficulty})")
            pronounce(f"The word is: {word}")
            pronounce(f"It means: {translation}")
            time.sleep(1)

# Quiz by category and difficulty
def quiz_user(category):
    if category not in flashcards:
        print(f"No flashcards found for category '{category}'.")
        return

    difficulty = input("Choose difficulty (easy, medium, or hard): ").lower()
    filtered_cards = [card for card in flashcards[category] if card[2] == difficulty]

    if not filtered_cards:
        print(f"No '{difficulty}' level flashcards in '{category}' category.")
        return

    try:
        count = int(input(f"How many {difficulty} {category} questions would you like to answer? "))
    except ValueError:
        print("Invalid number. Please try again.")
        return

    questions = random.sample(filtered_cards, min(count, len(filtered_cards)))

    for word, translation, _ in questions:
        pronounce(f"The word is: {word}")
        user_answer = input(f"What is the meaning of '{word}'? ").strip().lower()
        progress["total_attempts"] += 1

        similarity = difflib.SequenceMatcher(None, user_answer, translation.lower()).ratio()
        if similarity >= 0.7:
            print("✅ Correct!")
            progress["correct_answers"] += 1
        else:
            print(f"❌ Incorrect! The correct answer is: '{translation}'.")
            pronounce(f"The correct meaning of {word} is: {translation}")
            time.sleep(1)

# Plot progress
def plot_progress(correct, total):
    incorrect = total - correct
    labels = ['Correct', 'Incorrect']
    values = [correct, incorrect]
    colors = ['#4CAF50', '#F44336']

    fig, axs = plt.subplots(1, 2, figsize=(12, 5))
    fig.suptitle('Quiz Performance Overview', fontsize=16, fontweight='bold')

    axs[0].bar(labels, values, color=colors)
    axs[0].set_title('Answers Count')
    axs[0].set_ylabel('Number of Answers')
    axs[0].set_ylim(0, max(values) + 2)
    axs[0].grid(axis='y', linestyle='--', alpha=0.7)

    axs[1].pie(values, labels=labels, autopct='%1.1f%%', colors=colors, startangle=140, explode=(0.05, 0))
    axs[1].set_title('Correct vs Incorrect (%)')

    plt.tight_layout()
    plt.subplots_adjust(top=0.85)
    plt.show()

# Show statistics
def show_statistics():
    total = progress["total_attempts"]
    correct = progress["correct_answers"]
    accuracy = (correct / total * 100) if total > 0 else 0
    print(f"\n📈 Progress:\nTotal Attempts: {total}, Correct: {correct}, Accuracy: {accuracy:.2f}%")

    if total > 0:
        plot_progress(correct, total)

# Main menu
def main():
    while True:
        print("\n📚 Options:")
        print("1. Review flashcards")
        print("2. Take a quiz")
        print("3. Show progress")
        print("4. Exit")

        choice = input("Enter 1, 2, 3, or 4: ")

        if choice == '1':
            review_flashcards()
        elif choice == '2':
            print(f"Available categories: {', '.join(flashcards.keys())}")
            category = input("Enter a category to quiz on: ").lower()
            quiz_user(category)
        elif choice == '3':
            show_statistics()
        elif choice == '4':
            print("Goodbye! 👋 Keep practicing!")
            break
        else:
            print("Invalid choice. Please try again.")

# Start program
if __name__ == "__main__":
    main()
```

```python
import random
import time
import matplotlib.pyplot as plt
from gtts import gTTS
from IPython.display import Audio, display, clear_output
import os
import uuid
import difflib

# Global variables
reviewed_cards = []
reviewed_total = 0
quiz_count = 0

# Progress tracker
progress = {
    "total_attempts": 0,
    "correct_answers": 0
}

# Function to pronounce text using gTTS
def pronounce(text):
    try:
        filename = f"temp_{uuid.uuid4()}.mp3"
        tts = gTTS(text=text, lang='en')
        tts.save(filename)
        display(Audio(filename, autoplay=False))
        if os.path.exists(filename):
            os.remove(filename)
    except Exception as e:
        print(f"Pronunciation error: {e}")

# Flashcards categorized by topic
flashcards = {
    "vocabulary": [
        ('abundant', 'existing in large quantities', 'hard'),
        ('dull', 'boring', 'easy'),
        ('educated', 'learned', 'medium'),
        ('efficient', 'productive', 'medium'),
        ('fail', 'flunk', 'easy'),
        ('fast', 'quick', 'easy'),
        ('fresh', 'recent', 'easy'),
        ('gentle', 'mild', 'easy'),
        ('grow', 'expand', 'medium'),
        ('happy', 'joyful', 'easy'),
    ],
    "phrases": [
        ('break a leg', 'good luck', 'easy'),
        ('hit the nail on the head', 'get something exactly right', 'medium'),
        ('bit the bullet', 'endure a painful situation bravely', 'hard'),
        ('burn the midnight oil', 'work late into the night', 'medium'),
        ('kill two birds with one stone', 'solve two problems at once', 'easy'),
        ('under the weather', 'feeling sick', 'easy'),
        ('the ball is in your court', 'it is your turn to take action', 'easy'),
        ('through thick and thin', 'in good times and bad times', 'medium'),
        ('spill the beans', 'reveal a secret', 'easy'),
        ('once in a blue moon', 'something that happens rarely', 'easy'),
        ('put all your eggs in one basket', 'relying too much on one option', 'medium'),
        ("don't cry over spilled milk", "don't stress over something that can't be changed", 'hard'),
    ],
    "business": [
        ('ROI', 'Return on Investment', 'medium'),
        ('stakeholder', 'person with interest in a project', 'medium'),
    ]
}

# Review flashcards from selected categories
def review_flashcards():
    global reviewed_cards, reviewed_total, quiz_count
    reviewed_cards = []
    quiz_count = 0

    print(f"Available categories: {', '.join(flashcards.keys())}")
    categories_input = input("Enter one or more categories separated by commas: ").lower()
    selected_categories = [cat.strip() for cat in categories_input.split(",") if cat.strip() in flashcards]

    if not selected_categories:
        print("❌ No valid categories entered.")
        return

    try:
        count = int(input("How many flashcards in total would you like to review? "))
    except ValueError:
        print("❌ Invalid number.")
        return

    combined_cards = []
    for cat in selected_categories:
        combined_cards.extend([(cat, *card) for card in flashcards[cat]])
    random.shuffle(combined_cards)

    for i, (category, word, translation, difficulty) in enumerate(combined_cards[:count]):
        clear_output(wait=True)
        print(f"[{category.capitalize()}] {word:<35}: {translation:<40} ({difficulty})")
        pronounce(f"The word is: {word}")
        pronounce(f"It means: {translation}")
        reviewed_cards.append((word, translation))
        time.sleep(2)

    reviewed_total = len(reviewed_cards)

# Quiz only from reviewed flashcards
def quiz_from_reviewed():
    global quiz_count
    if not reviewed_cards:
        print("❌ You haven't reviewed any flashcards yet.")
        return

    try:
        count = int(input(f"You reviewed {len(reviewed_cards)} flashcards. How many do you want to be quizzed on? "))
    except ValueError:
        print("❌ Invalid number.")
        return

    count = min(count, len(reviewed_cards))
    quiz_set = random.sample(reviewed_cards, count)

    for word, translation in quiz_set:
        pronounce(f"The word is: {word}")
        user_answer = input(f"What is the meaning of '{word}'? ").strip().lower()
        progress["total_attempts"] += 1

        similarity = difflib.SequenceMatcher(None, user_answer, translation.lower()).ratio()
        if similarity >= 0.7:
            print("✅ Correct!")
            progress["correct_answers"] += 1
        else:
            print(f"❌ Incorrect! The correct answer is: '{translation}'.")
            pronounce(f"The correct meaning of {word} is: {translation}")
        time.sleep(1)

    quiz_count += count

# Plot accuracy progress
def plot_accuracy_progress(correct, total):
    incorrect = total - correct
    labels = ['Correct', 'Incorrect']
    values = [correct, incorrect]
    colors = ['#4CAF50', '#F44336']

    fig, axs = plt.subplots(1, 2, figsize=(12, 5))
    fig.suptitle('Quiz Performance Overview', fontsize=16, fontweight='bold')

    axs[0].bar(labels, values, color=colors)
    axs[0].set_title('Answers Count')
    axs[0].set_ylabel('Number of Answers')
    axs[0].set_ylim(0, max(values) + 2)
    axs[0].grid(axis='y', linestyle='--', alpha=0.7)

    axs[1].pie(values, labels=labels, autopct='%1.1f%%', colors=colors, startangle=140, explode=(0.05, 0))
    axs[1].set_title('Correct vs Incorrect (%)')

    plt.tight_layout()
    plt.subplots_adjust(top=0.85)
    plt.show()

# Plot learning progress
def plot_learning_progress():
    learned = quiz_count
    remaining = reviewed_total - quiz_count
    labels = ['Learned (Quizzed)', 'Remaining']
    values = [learned, max(0, remaining)]
    colors = ['#2196F3', '#FFC107']

    plt.figure(figsize=(6, 5))
    plt.bar(labels, values, color=colors)
    plt.title('Flashcard Learning Progress', fontsize=14, fontweight='bold')
    plt.ylabel('Number of Flashcards')
    plt.ylim(0, reviewed_total + 1)
    for i, v in enumerate(values):
        plt.text(i, v + 0.1, str(v), ha='center', fontweight='bold')
    plt.grid(axis='y', linestyle='--', alpha=0.5)
    plt.tight_layout()
    plt.show()

# Show statistics
def show_statistics():
    total = progress["total_attempts"]
    correct = progress["correct_answers"]
    accuracy = (correct / total * 100) if total > 0 else 0

    print(f"\n📈 Quiz Accuracy:\nTotal Attempts: {total}, Correct: {correct}, Accuracy: {accuracy:.2f}%")
    if total > 0:
        plot_accuracy_progress(correct, total)

    if reviewed_total > 0:
        print(f"\n📊 Learning Progress:\nReviewed: {reviewed_total}, Quizzed: {quiz_count}, Remaining: {reviewed_total - quiz_count}")
        plot_learning_progress()

# Main menu
def main():
    while True:
        print("\n📚 Options:")
        print("1. Review flashcards")
        print("2. Quiz from reviewed flashcards")
        print("3. Show progress")
        print("4. Exit")

        choice = input("Enter 1, 2, 3, or 4: ")

        if choice == '1':
            review_flashcards()
        elif choice == '2':
            quiz_from_reviewed()
        elif choice == '3':
            show_statistics()
        elif choice == '4':
            print("Goodbye! 👋 Keep practicing!")
            break
        else:
            print("Invalid choice. Please try again.")

# Start program
main()
```

```python
import random
import time
import matplotlib.pyplot as plt
from gtts import gTTS
from IPython.display import Audio, display, clear_output
import os
import uuid
import difflib

# Global variables
reviewed_cards = []
reviewed_total = 0
quiz_count = 0

# Progress tracker
progress = {
    "total_attempts": 0,
    "correct_answers": 0
}

# Function to pronounce text using gTTS
def pronounce(text):
    try:
        filename = f"temp_{uuid.uuid4()}.mp3"
        tts = gTTS(text=text, lang='en')
        tts.save(filename)
        display(Audio(filename, autoplay=False))
        if os.path.exists(filename):
            os.remove(filename)
    except Exception as e:
        print(f"Pronunciation error: {e}")

# Flashcards categorized by topic
flashcards = {
    "vocabulary": [
        ('abundant', 'existing in large quantities', 'hard'),
        ('dull', 'boring', 'easy'),
        ('educated', 'learned', 'medium'),
        ('efficient', 'productive', 'medium'),
        ('fail', 'flunk', 'easy'),
        ('fast', 'quick', 'easy'),
        ('fresh', 'recent', 'easy'),
        ('gentle', 'mild', 'easy'),
        ('grow', 'expand', 'medium'),
        ('happy', 'joyful', 'easy'),
    ],
    "phrases": [
        ('break a leg', 'good luck', 'easy'),
        ('hit the nail on the head', 'get something exactly right', 'medium'),
        ('bit the bullet', 'endure a painful situation bravely', 'hard'),
        ('burn the midnight oil', 'work late into the night', 'medium'),
        ('kill two birds with one stone', 'solve two problems at once', 'easy'),
        ('under the weather', 'feeling sick', 'easy'),
        ('the ball is in your court', 'it is your turn to take action', 'easy'),
        ('through thick and thin', 'in good times and bad times', 'medium'),
        ('spill the beans', 'reveal a secret', 'easy'),
        ('once in a blue moon', 'something that happens rarely', 'easy'),
        ('put all your eggs in one basket', 'relying too much on one option', 'medium'),
        ("don't cry over spilled milk", "don't stress over something that can't be changed", 'hard'),
    ],
    "business": [
        ('ROI', 'Return on Investment', 'medium'),
        ('stakeholder', 'person with interest in a project', 'medium'),
    ]
}

# Review flashcards from selected categories
def review_flashcards():
    global reviewed_cards, reviewed_total, quiz_count
    reviewed_cards = []
    quiz_count = 0

    print(f"Available categories: {', '.join(flashcards.keys())}")
    categories_input = input("Enter one or more categories separated by commas: ").lower()
    selected_categories = [cat.strip() for cat in categories_input.split(",") if cat.strip() in flashcards]

    if not selected_categories:
        print("❌ No valid categories entered.")
        return

    try:
        count = int(input("How many flashcards in total would you like to review? "))
    except ValueError:
        print("❌ Invalid number.")
        return

    combined_cards = []
    for cat in selected_categories:
        combined_cards.extend([(cat, *card) for card in flashcards[cat]])
    random.shuffle(combined_cards)

    for i, (category, word, translation, difficulty) in enumerate(combined_cards[:count]):
        clear_output(wait=True)
        print(f"[{category.capitalize()}] {word:<35}: {translation:<40} ({difficulty})")
        pronounce(f"The word is: {word}")
        pronounce(f"It means: {translation}")
        reviewed_cards.append((word, translation))
        time.sleep(2)

    reviewed_total = len(reviewed_cards)

# Quiz only from reviewed flashcards
def quiz_from_reviewed():
    global quiz_count
    if not reviewed_cards:
        print("❌ You haven't reviewed any flashcards yet.")
        return

    try:
        count = int(input(f"You reviewed {len(reviewed_cards)} flashcards. How many do you want to be quizzed on? "))
    except ValueError:
        print("❌ Invalid number.")
        return

    count = min(count, len(reviewed_cards))
    quiz_set = random.sample(reviewed_cards, count)

    for word, translation in quiz_set:
        pronounce(f"The word is: {word}")
        user_answer = input(f"What is the meaning of '{word}'? ").strip().lower()
        progress["total_attempts"] += 1

        similarity = difflib.SequenceMatcher(None, user_answer, translation.lower()).ratio()
        if similarity >= 0.7:
            print("✅ Correct!")
            progress["correct_answers"] += 1
        else:
            print(f"❌ Incorrect! The correct answer is: '{translation}'.")
            pronounce(f"The correct meaning of {word} is: {translation}")
        time.sleep(1)

    quiz_count += count

# Plot accuracy progress
def plot_accuracy_progress(correct, total):
    incorrect = total - correct
    labels = ['Correct', 'Incorrect']
    values = [correct, incorrect]
    colors = ['#4CAF50', '#F44336']

    fig, axs = plt.subplots(1, 2, figsize=(12, 5))
    fig.suptitle('Quiz Performance Overview', fontsize=16, fontweight='bold')

    axs[0].bar(labels, values, color=colors)
    axs[0].set_title('Answers Count')
    axs[0].set_ylabel('Number of Answers')
    axs[0].set_ylim(0, max(values) + 2)
    axs[0].grid(axis='y', linestyle='--', alpha=0.7)

    axs[1].pie(values, labels=labels, autopct='%1.1f%%', colors=colors, startangle=140, explode=(0.05, 0))
    axs[1].set_title('Correct vs Incorrect (%)')

    plt.tight_layout()
    plt.subplots_adjust(top=0.85)
    plt.show()

# Plot learning progress bar (reviewed vs quizzed)
def plot_learning_progress():
    learned = quiz_count
    remaining = reviewed_total - quiz_count
    labels = ['Learned (Quizzed)', 'Remaining']
    values = [learned, max(0, remaining)]
    colors = ['#2196F3', '#FFC107']

    plt.figure(figsize=(6, 5))
    plt.bar(labels, values, color=colors)
    plt.title('Flashcard Learning Progress', fontsize=14, fontweight='bold')
    plt.ylabel('Number of Flashcards')
    plt.ylim(0, reviewed_total + 1)
    for i, v in enumerate(values):
        plt.text(i, v + 0.1, str(v), ha='center', fontweight='bold')
    plt.grid(axis='y', linestyle='--', alpha=0.5)
    plt.tight_layout()
    plt.show()

# NEW: Overall learning summary bar chart
def plot_overall_learning_summary():
    labels = ['Reviewed', 'Attempted (Quiz)', 'Correct Answers']
    values = [reviewed_total, quiz_count, progress['correct_answers']]
    colors = ['#2196F3', '#FFC107', '#4CAF50']

    plt.figure(figsize=(7, 5))
    bars = plt.bar(labels, values, color=colors)
    plt.title('Overall Learning Summary', fontsize=14, fontweight='bold')
    plt.ylabel('Number of Words')
    plt.ylim(0, max(values) + 2)

    for bar in bars:
        yval = bar.get_height()
        plt.text(bar.get_x() + bar.get_width() / 2, yval + 0.1, f'{int(yval)}', ha='center', va='bottom', fontweight='bold')

    plt.grid(axis='y', linestyle='--', alpha=0.5)
    plt.tight_layout()
    plt.show()

# Show statistics
def show_statistics():
    total = progress["total_attempts"]
    correct = progress["correct_answers"]
    accuracy = (correct / total * 100) if total > 0 else 0

    print(f"\n📈 Quiz Accuracy:\nTotal Attempts: {total}, Correct: {correct}, Accuracy: {accuracy:.2f}%")
    if total > 0:
        plot_accuracy_progress(correct, total)

    if reviewed_total > 0:
        print(f"\n📊 Learning Progress:\nReviewed: {reviewed_total}, Quizzed: {quiz_count}, Remaining: {reviewed_total - quiz_count}")
        plot_learning_progress()
        plot_overall_learning_summary()  # ✅ New graph here

# Main menu
def main():
    while True:
        print("\n📚 Options:")
        print("1. Review flashcards")
        print("2. Quiz from reviewed flashcards")
        print("3. Show progress")
        print("4. Exit")

        choice = input("Enter 1, 2, 3, or 4: ")

        if choice == '1':
            review_flashcards()
        elif choice == '2':
            quiz_from_reviewed()
        elif choice == '3':
            show_statistics()
        elif choice == '4':
            print("Goodbye! 👋 Keep practicing!")
            break
        else:
            print("Invalid choice. Please try again.")

# Start program
main()
```
